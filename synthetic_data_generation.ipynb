```python
import cv2
import numpy as np
```

```python
# Output video settings
W, H = 128, 128
FPS = 60
DURATION = 5  # seconds
FRAMES = FPS * DURATION

# Drumming motion settings
amp = 25              # vertical amplitude
freq = 4              # Hz (drumming speed)
phase_shift = np.pi   # left hand opposite of right hand
radius = 8            # "hand" circle size

# Hand positions (horizontal)
left_x = 40
right_x = 88
y_center = 64

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter("synthetic_airdrums.mp4", fourcc, FPS, (W, H), isColor=False)

for t in range(FRAMES):
    frame = np.zeros((H, W), dtype=np.uint8)

    time = t / FPS

    # Vertical positions (sinusoidal)
    left_y  = int(y_center + amp * np.sin(2*np.pi*freq*time))
    right_y = int(y_center + amp * np.sin(2*np.pi*freq*time + phase_shift))

    # Draw hands
    cv2.circle(frame, (left_x, left_y),  radius, 255, -1)
    cv2.circle(frame, (right_x, right_y), radius, 255, -1)

    out.write(frame)

out.release()
print("Saved synthetic_airdrums.mp4")
```
